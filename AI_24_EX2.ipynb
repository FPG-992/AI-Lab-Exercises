Σε αυτό το σημείο θα γίνει μια αναβάθμιση του συστήματος έτσι ώστε να παράγονται καλύτερες  συστάσεις οι όποιες θα λαμβάνουν υπόψιν και τις προτιμήσεις του χρήστη, οι όποιες θα εξάγονται από τις αξιολογήσεις που έχει κάνει μέχρι στιγμής.

Η εκπαίδευση του recommender θα γίνεται ως εξής:

Για κάθε ταινία θα υπάρχει ένα score το όποιο αρχικά θε είναι ίσο με 0 και θα διαμορφώνεται από τις αξιολογήσεις κάθε user. Έτσι για έναν χρήστη με βάση τις αξιολογήσεις που υπάρχουν στο αρχείο train_ratings θα πρέπει:

1.	Για κάθε ταινία που έχει βαθμολογήσει να βρίσκονται οι κοινές ταινίες ανά κλίμακα και στο μέχρι τώρα σκορ κάθε παρόμοιας ταινίας θα προστίθεται ένα βάρος το όποιο θα μπορούσε να είναι το ποσοστό ομοιότητας της ταινίας (δηλαδή ένα βάρος για κάθε κλίμακα, αν δύο ταινίες μοιάζουν ενισχύουμε το βάρος που προσθέτουμε από το να μοιάζουν λιγότερο) επί τον βαθμό που έχει βάλει ο χρήστης για την αρχική ταινία (διαφορετικό είναι ο χρήστης να έχει βάλει 5/5 ή 1/5 σε μια ταινία από 3/5).

2. Στη συνέχεια, ανάλογα με το σκορ που έχει σχηματιστεί για κάθε ταινία, θα επιλέγεται αν αυτή θα μπορούσε να είναι προτεινόμενη για τον χρήστη ή όχι και θα μετράμε το πόσο καλά τα πήγε το σύστήμά μας με βάση ορισμένες μετρικές.

Η λογική πίσω από την παραπάνω διαδικασία είναι ότι παρόμοιες ταινίες θα έχουν ανάλογο βαθμό. Για παράδειγμα αν ένας χρήστης έχει αξιολογήσει αρκετές  ταινίες οι όποιες είναι sci-fiction με 5/5 τότε μια ταίνια sci-fiction την όποια δεν έχει δεί λογικά θα του αρέσει και θα έπρεπε να την προτείνουμε.

Μετά την εκπαίδευση του συστήματος σας καλείστε να δοκιμάσετε τον recommender που κατασκευάσατε στην πράξη. Για τον σκοπό αυτό θα φορτώσετε το αρχείο test_ratings.csv όπου περιέχονται οι αξιολογήσεις του ίδιου χρήστη για άλλες ταινίες. Το σύστημά σας πρέπει να προβλέπει αν μια ταινία θα πρέπει να προταθεί στον χρήστη. Μια ταίνια θεωρούμε οτι προτείνεται στον  χρήστη, αν έχει βαθμό μεγαλύτερο του 3 στο test_ratings. Συνεπώς για την επίβλεψη του συστήματός σας θα πρέπει για κάθε μια από τις ταινίες του αρχείου test_ratings.csv να προβλέψετε εάν θα αρέσει στο χρήστη ή όχι ώστε να την προτείνετε.

ΠΡΟΣΟΧΗ! Η "βαθμολογία" που θα υπολογίσετε για κάθε ταινία κατά το training δεν είναι απαραίτητα πρόβλεψη της βαθμολογίας που θα έβαζε ο χρήστης.

Στην συνέχεια, σε συνδιασμό με τις πραγματικές απαντήσεις του χρήστη θα αξιολογήσετε το σύστημά σας χρησιμοποιόντας τις μετρικές: precision, recall, f1 οι οποίες είναι οι πλέον γνωστές μετρικές και ευρέως χρησιμοποιούμενες τεχνικές για την επίβλεψη-μέτρηση απόδοσης ανάλογων συστημάτων.

1. Precision: Δείχνει πόσο ακριβές είναι το σύστημα. Υπολογίζει πόσα από τα στιγμιότυπα τα όποια προβλέψαμε ότι ανήκουν σε μια κλάση όντως ανήκουν σε αυτή. Η μετρική αυτή μας δίνει μια εικόνα σχετικά με τον αριθμό των ταινίων που προβλέψαμε ως προτεινόμενες ενώ δεν θα έπρεπε.

2. Recall: Υπολογίζει πόσα από τα στιγμιότυπα που ανήκουν σε μια κλάση (π.χ. προτεινόμενες ταινίες) προβλέφθηκαν σωστά.

3. F1: Είναι ένας μέσος μεταξύ των παραπάνω δυο μετρικών, έτσι ώστε να διατηρείται μια ισορροπία μεταξύ τους. Υπολογίζεται από την παρακάτω σχέση:

$$F_1=2\frac{Precision\times{Recall}}{Precision+Recall}$$

Οι παραπάνω συναρτήσεις παρέχονται από την βιβλιοθήκη scikit-learn.

Τέλος για την καλύτερη επίβλεψη του συστήματος σας μπορείτε να εκπαιδεύσετε  τον recommender σας με ένα υποσύνολο ταινίων από ελάχιστες, λίγες μέχρι και πολλές (π.χ. 3, 5, 10, 50, ...) για να μελετήσετε κατά πόσο σας βοηθούν οι επιπλέον αξιολογήσεις κάθε φορά (δηλαδή κατά πόσο βελτιώνονται οι παραπάνω μετρικές στο test set).

Έτσι π.χ. μπορείτε να εντοπίσετε περιπτώσεις όπως για παράδειγμα ότι με έναν recommender μπορεί να μην επιτυγχάνετε πολύ υψηλό σκορ όσο με άλλους, αλλά το βέλτιστο σκόρ σας επιτυγχάνεται πολύ γρηγόρα π.χ. με μόνο 10 ταινίες αντί 100. # **Μέρος 1: Μελέτη των Metadata, Δημιουργία κόσμου και των βασικών queries.**


Κοιτάμε το αρχείο μας (movies_metadata.csv) και βλέπουμε πως τα row που μας ενδιαφέρουν είναι movie_title, language, vote_average, με το οποίο κάνουμε classify την ταινία με βάση την βαθμολογία, title_year, director_name, plot_keywords, genres, production companies, production countries και κοιτάμε επίσης αν είναι ασπρόμαυρη η ταινία αν έχει το keyword black and white ως plot keyword.

Μετά παίρνουμε τα δεδομένα αυτά αφού τα καθαρίσουμε με την clean_text και τα προσθέτουμε ως fact αν δεν είναι UNK το movie title.

Εδώ να σημειώσουμε πως είχαμε μεγάλο θέμα με το να πάρουμε τα δεδομένα από τα production companies μιας και είχαν special characters που πετούσαν error το οποίο κάναμε handle παρακάτω.

Τέλος κάναμε sort τα facts καθώς έτσι τα ζητάει η prolog (literals) και τα αποθηκεύσαμε σε μία βάση με όνομα facts.pl.

````python
import pandas as pd
from pyswip import Prolog
import csv
import json

# Η βιβλιοθήκη pandas είναι χρήσιμη για την εργασία με τέτοια δεδομένα
# path = '/content/drive/MyDrive/'
# Διάβασμα του αρχείου 'movie_metadata.csv'
data = pd.read_csv("movies_metadata.csv")
# Στο csv υπαρχούν κελία με nan τιμές
# Στις θέσεις αυτές βάζουμε 'UNK' πράγμα που  κάνουμε με την παρακάτω συνάρτηση
data.fillna("UNK", inplace=True)


# Preview the first 5 lines of the loaded data
# data.head()
def clean_text(text):
    text = text.replace("\xa0", "")
    text = text.replace("'", "")
    return text


# create World
# Ορίζουμε τον κόσμο μας
prolog = Prolog()

# Για κάθε row του πίνακα φτιάχνουμε τα κατηγορήματα που θέλουμε να αποθηκέυσουμε
# αρχικά σε μια λίστα με το όνομα literals
literals = []
movie_score = {}
decode_errors = 0
color = "Color"  # default value
for row in data.itertuples(index=True, name="Pandas"):
    movie_title = clean_text(getattr(row, "movie_title"))
    language = clean_text(getattr(row, "language"))
    vote_average = getattr(row, "vote_average")
    title_year = getattr(row, "title_year")  # θέλω την δεκαετία
    director = clean_text(getattr(row, "director_name"))
    plot_keywords = clean_text(getattr(row, "plot_keywords"))

    # εάν το όνομα της ταινίας είναι 'UNK' τότε δεν την προσθέτουμε στον κόσμο
    if movie_title != "UNK":
        if language != "UNK":
            literals.append("language('" + movie_title + "','" + language + "')")
        if vote_average != "UNK":
            if vote_average >= 7:
                literals.append("score('" + movie_title + "','high')")
            elif vote_average >= 4.5 and vote_average < 7:
                literals.append("score('" + movie_title + "','average')")
            else:
                literals.append("score('" + movie_title + "','low')")
        if title_year != "UNK":
            decade = int((title_year // 10) * 10)
            literals.append("decade('" + movie_title + "','" + str(decade) + "')")
        if director != "UNK":
            literals.append("director('" + movie_title + "','" + director + "')")

        for genre in getattr(row, "genres").split("|"):
            literals.append("genre('" + movie_title + "','" + genre + "')")

        countries = json.loads(getattr(row, "production_countries").replace("'", '"'))
        country_names = [country["name"] for country in countries]
        # print the movie and each of its country name seperately
        for country in country_names:
            literals.append("country('" + movie_title + "','" + country + "')")

        # resolving decode error because of , inside names use of quotes, commas and other special characters
        try:
            # Retrieve the JSON string and replace single quotes with double quotes
            json_string = getattr(row, "production_companies").replace("'", '"')
            # print("Debug - JSON String:", json_string)  # Print the JSON string for debugging

            # Load the JSON to see if it parses without error
            production_studios = json.loads(json_string)
            production_studio_names = [studio["name"] for studio in production_studios]

            # Output each studio and movie title
            for studio in production_studio_names:
                literals.append(
                    "production_company('" + movie_title + "','" + studio + "')"
                )
                # print("production_studio('"+ movie_title +"','"+ studio +"')")

        except json.JSONDecodeError as e:
            decode_errors += 1
            # Αυτό συμβαίνει λόγω των " που υπάρχουν στα ονόματα εταιρειών παραγωγής. Το θεωρούμε UNK
        except Exception as general_error:
            print("An unexpected error occurred:", general_error)

        for plot_keyword in plot_keywords.split("|"):
            literals.append(
                "plot_keyword('" + movie_title + "','" + plot_keyword + "')"
            )
            # print("keyword('"+ movie_title +"','"+ plot_keyword +"')")
            if plot_keyword == "black and white":
                color = "black_and_white"
        # print("color('"+ movie_title +"','" + color + "')")
        literals.append("color('" + movie_title + "','" + color + "')")

        for actor in [
            getattr(row, "actor_1_name"),
            getattr(row, "actor_2_name"),
            getattr(row, "actor_3_name"),
        ]:
            if actor != "UNK":  # Check if actor name is not 'UNK'
                cleaned_actor = actor.replace(
                    "'", "\\'"
                )  # Replace single quotes with escape sequence
                literals.append(f"actor('{movie_title}','{cleaned_actor}')")
                # print(f"actor('{movie_title}','{cleaned_actor}')")


# print (decode_errors)
# Η Prolog θέλει τα κατηγορήματά της με την σειρά
literals.sort()
for literal in literals:
    prolog.assertz(literal)
    # print (literal +'.')

with open("facts.pl", "w") as f:
    for literal in literals:
        f.write(literal + ".\n")

# Επίσης μπορούμε να κάνουμε consult ένα έτοιμο αρχείο στον κόσμο όπως παρακάτω
# prolog.consult("facts.pl")
````

Εδώ βρίσκονται τα predicates που φτιάξαμε τα οποία συμπεριλαμβάνουν όλα τα ζητούμενα:

Δοκιμάζουμε το Recommender μας 

Κάνουμε μια δοκιμή του recommender μας για να δούμε αν όλα είναι λειτουργικά (Debugging)

````python
def simple_recommender(movie):
    def check(query):
        s = set()
        for solution in query:
            m = solution["M"]
            if m not in s:
                s.add(m)
        return s

    # Initialize answer dictionary
    answers = {}

    # Check for each level of similarity, from 5 to 1
    for level in range(5, 0, -1):
        query = prolog.query(f"similarity_{level}('{movie}', M)")
        s = check(query)
        if s:
            answers[level] = list(s)
        query.close()

    # Return the answer dictionary
    return answers


# Example usage
movie_name = "The Twilight Saga: Breaking Dawn"
recommended_movies = simple_recommender(movie_name)
````

Για ένα subset ταινιών θα κάνουμε ένα δοκιμαστικό run

````python
def test_run(test_run_movies, max_recommended=3):
    for movie in test_run_movies:
        recommended_movies = simple_recommender(movie)
        print(f"\nRecommended Movies for {movie}:")
        for level in range(5, 0, -1):
            if level in recommended_movies:
                print(
                    f"Similarity Level {level}: {recommended_movies[level][:max_recommended]}"
                )


# Example usage
test_run_movies = [
    "Avatar",
    "Harry Potter and the Chamber of Secrets",
    "Superman",
    "The Dark Knight",
    "The Avengers",
    "The Lord of the Rings: The Return of the King",
    "The Shawshank Redemption",
    "The Godfather",
]
test_run(test_run_movies, max_recommended=3)
````

# **3ο Μέρος: Recommendation System Με βάση τις προτιμήσεις-Αξιολογήσεις του χρήστη-Εκπαίδευση και Πρόβλεψη**


Αρχικά μελετάμε τις αξιολογήσεις κάθε χρήστη για να καταλάβουμε την δομή και τις πληροφορίες κάθε αρχείου.




Στο σημείο αυτό ξανακάνουμε redefine τον recommender για να μπορέσουμε να τον κάνουμε integrate με την train_recommender.

Σε προηγούμενες μας δοκιμές είχαμε πολύ μεγάλο θέμα με τα αρνητικά βάρη και αναπροσαρμόσαμε και τα rating weights και τα score weights.

````python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
import random

train_ratings = pd.read_csv("train_ratings.csv")
test_ratings = pd.read_csv("test_ratings.csv")

rating_weights = {0: 0.1, 1: 0.2, 2: 0.4, 3: 0.6, 4: 0.8, 5: 1}
score_weights = {
    1: 0.2,
    2: 0.4,
    3: 0.6,
    4: 0.8,
    5: 1,
}  # ανάλογα με τα επίπεδα ομοιότητας που έχουν οριστεί στην simple_recommender


def find_movies(movie, similarity):
    q = prolog.query("similarity_" + similarity + "('" + movie + "',M)")
    ans = []
    for soln in q:
        ans.append((soln["M"], similarity))
    return set(ans)


def simple_recommender(movie):
    ans = set()
    for i in range(5, 0, -1):
        ans.update(find_movies(movie, str(i)))
    return list(ans)


def train_recommender(ratings, rating_weights, score_weights, number_of_movies):
    if number_of_movies > len(ratings):
        number_of_movies = len(ratings)

    if number_of_movies != -1:
        indexes = random.sample(range(len(ratings)), number_of_movies)
        ratings = ratings.iloc[indexes]

    movie_score = {}
    for row in tqdm(ratings.itertuples(index=True, name="Pandas")):
        movie = clean_text(getattr(row, "movie_title"))
        rating = getattr(row, "rating")
        similar_movies = simple_recommender(movie)

        for similar_movie, similarity in similar_movies:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = (
                    rating_weights[int(rating)] * score_weights[int(similarity)]
                )
            else:
                movie_score[similar_movie] += (
                    rating_weights[int(rating)] * score_weights[int(similarity)]
                )
            # print(f"Processed {movie}: Similar {similar_movie} (sim {similarity}) with score {movie_score[similar_movie]}")
    return movie_score


# Αυτό είναι ένα παράδειγμα για το πως θα μπορούσε να υλοποιήθει η predict.
# Έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > 0.
def predict_example(ratings, movie_score):
    real, pred = [], []
    for i, row in enumerate(ratings.itertuples(index=True, name="Pandas")):
        movie = clean_text(getattr(row, "movie_title"))
        rating = getattr(row, "rating")

        if movie in movie_score:  # αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(
                int(movie_score[movie] > 0)
            )  # heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(
                int(rating > 3)
            )  # έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            # η συνθήκη αυτή δεν μπορεί να αλλάξει
        else:  # δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))

    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred)
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
````
